```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

'''
Goalkeeper :  69 in dataset
Defender :  172 in dataset
Midfield :  142 in dataset
Attacking :  172 in dataset
'''
def avg_position(position, column):
  return round(bundesliga[bundesliga['position'].str.contains(
      position, case=False)].groupby('position')[column].mean(),1)


# Read the csv with bundesliga data
bundesliga = pd.read_csv('bundesliga_player.csv')

# Avoid n/a values
bundesliga = bundesliga.fillna('No data')

# Getting all the heigth means by position
gkp_avg = avg_position('Goalkeeper', 'height')
def_avg = avg_position('Defender','height')
mid_avg = avg_position('Midfield', 'height')
att_avg = avg_position('Attack', 'height')


# Getting all the age means by position
gkp_age_avg = avg_position('Goalkeeper', 'age')
def_age_avg = avg_position('Defender','age')
mid_age_avg = avg_position('Midfield', 'age')
att_age_avg = avg_position('Attack', 'age')


# Concat all the means in a new dataframe
height_pos = pd.concat([gkp_avg, def_avg,mid_avg, att_avg ]).to_dict()

age_pos = pd.concat([gkp_age_avg, def_age_avg,mid_age_avg, att_age_avg ]).to_dict()

# Getting keys and values for bar chart
x_values = height_pos.keys()
y_values = height_pos.values()

# Creating the fig to show the map
fig, ax = plt.subplots()
# Create a colormap
cmap = plt.cm.plasma  # You can choose any colormap
# Normalize the range for colormap
colors = [cmap(i / len(y_values)) for i in range(len(y_values))]
bar_labels = ax.bar(x_values, y_values, color=colors, width=0.7)
# Each bar with values
# Chart titles
ax.bar_label(bar_labels)
plt.title('Average height by position')
plt.xlabel('Position')
plt.ylabel('Height (cm)')
# Rotation for x labels to avoid overlap
plt.xticks(rotation=90)
plt.show()
```

```python
# Age average chart

x_age_values = age_pos.keys()
y_age_values = age_pos.values()

# Creating the fig to show the map
fig, ax = plt.subplots()
# Create a colormap
cmap = plt.cm.cividis  # You can choose any colormap
# Normalize the range for colormap
colors_age = [cmap(i / len(y_age_values)) for i in range(len(y_age_values))]
bar_age_labels = ax.bar(x_age_values, y_age_values, color=colors_age, width=0.6)
# Each bar with values
# Chart titles
ax.bar_label(bar_age_labels)
plt.title('Average age by position')
plt.xlabel('Position')
plt.ylabel('Age (years)')
# Rotation for x labels to avoid overlap
plt.xticks(rotation=90)
plt.show()
```
